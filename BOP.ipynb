```python
Data=[]
D_source = 'International Monetary Fund (IMF)'
D_providerID= 2
for num,Ind_code in enumerate(Indicators):
    url = f'http://dataservices.imf.org/REST/SDMX_JSON.svc/CompactData/BOP/A..{Ind_code}.?startPeriod=2010&endPeriod={y}'
    a = 0
    print(f"Indicator {num}:",Ind_code)
    data = None
    while data is None:
        try:
            a +=1
            print("\t atempt:",a)
            data=(requests.get(url).json()['CompactData']['DataSet']['Series'])
        except:
            time.sleep(3)
            continue
    Ind_name = Indicators[Ind_code][0]
    S_source= f'Retrived from: {url}, On: {str(D.date())}, By web scraping using the python libraries: requests and pandas'
    
    Mode_code=Indicators[Ind_code][1]
    Unit_code=Indicators[Ind_code][2]
    
    for i in range(len(data)):
        if 'Obs' in data[i]:
            obs = data[i]['Obs']
            area_code = data[i].get('@REF_AREA')
            scale = str(data[i].get('@UNIT_MULT'))
            Unit_Mult = Scales[scale][0]
            
            if area_code in Area_Code:
                area_name = Area_Code[area_code][0]
                area_c = Area_Code[area_code][1]
                for o in obs:
                    if o.get('@OBS_VALUE') is not None:
                        o_v = float(o.get('@OBS_VALUE'))
                        cal_v = int(o_v*10**Scales[scale][1])
                        Data.append([area_c, area_name,\
                                     D_source, D_providerID,\
                                     S_source, Ind_code,\
                                     Ind_name, Mode_code,
                                     Unit_code, scale,\
                                     o.get('@TIME_PERIOD'),\
                                     o_v, cal_v])
                        
DataFrame=pd.DataFrame(Data,columns=['Areacode','Areaenglish','DataSource','DataProviderId',\
                                     'SpecificSource','Indicatorcode','Indicatorname','Modecode',\
                                     'Unitcode','Scale','Year','Value','cal_value'])
DataFrame.to_excel('BOP.xlsx', index=False)
print('All Done!')    
```
