```python
###### Input files ############################################################
# Specify the path of the folder containing the BDDS files:
path = './OPRI/'

# Create a ZipFile Object and load OPRI.zip in it
with ZipFile('OPRI.zip', 'r') as zipObj:
    # Extract all the contents of zip file into OPRI folder
    zipObj.extractall(path)
    zipObj.close()

#Remove the zip file and files inside OPRI folder except the file OPRI_DATA_NATIONAL.csv
os.remove('OPRI.zip') 
os.remove(path+'OPRI_METADATA.csv') 
os.remove(path+'OPRI_COUNTRY.csv') 
os.remove(path+'OPRI_LABEL.csv') 
os.remove(path+'OPRI_README_RELEASE_2022_March.md') 
os.remove(path+'OPRI_DATA_REGIONAL.csv') 
os.remove(path+'OPRI_REGION.csv') 
```

```python
# Read the data set to memory
dataSet = pd.read_csv(path+'OPRI_DATA_NATIONAL.csv')     #Data file
```

```python
#Changing numeral indicators to string and extracting unique values
dataSet["INDICATOR_ID"] = dataSet["INDICATOR_ID"].astype(str)
#Defining the list of year, country and indicator of interest
countryList = np.array(countries)
indicatorList = np.array(selected_Indi)
yearList = np.sort(dataSet["YEAR"].unique())

aSubset = dataSet[(dataSet['YEAR'].isin(yearList)) &\
                      (dataSet['COUNTRY_ID'].isin(countryList)) &\
                          (dataSet['INDICATOR_ID'].isin(indicatorList))]
```

```python
Data=[]
D_source = 'United Nations Educational, Scientific and Cultural Organization (UNESCO)'
D_providerID= 13
S_source= ''#f'Retrived from: {url}, On: {str(D.date())}, By web scraping using the python libraries: requests and pandas'
for i in range(len(aSubset)):
    ind_c = aSubset.iloc[i,0]
    
    ECA_ind_c = Indicators[ind_c][0]
    area_c = aSubset.iloc[i,1]
    c_name = Area_Code[area_c]
    ind_name = Indicators[ind_c][1]
    m_code=''
    u_code=Indicators[ind_c][2]
    
    atributes = ind_name.split()
    
    scale=''
    
    year = aSubset.iloc[i,2]
    val = aSubset.iloc[i,3]
    if atributes[-1][1:-1]=='millions':
        scale='millions'    
        c_val = 1e6*val
    else:
        c_val = val
    
    Data.append([area_c,c_name,D_source,D_providerID,S_source,ECA_ind_c,\
                ind_name,m_code,u_code,scale,year,val,c_val])
    
    
DataFrame=pd.DataFrame(Data,columns=['Areacode','Areaenglish','DataSource',\
                                     'DataProviderId','SpecificSource','Indicatorcode',\
                                     'Indicatorname','Modecode','Unitcode','Scale',\
                                     'Year','Value','cal_value'])

DataFrame = DataFrame.sort_values(by=['Areaenglish'], ascending=True)
DataFrame.to_excel('Education.xlsx', index=False)
os.remove(path+'OPRI_DATA_NATIONAL.csv') 
os.rmdir(path)
print('All Done!')    
```
