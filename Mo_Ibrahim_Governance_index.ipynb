```python
# Create an URL object
url = 'https://iiag.online/data.html?meas=GOVERNANCE&loc=AO|BF|BI|BJ|BW|CD|CF|CG|CI|CM|CV|DJ|DZ|EG|ER|ET|GA|GH|GM|GN|GQ|GW|KE|KM|LR|LS|LY|MA|MG|ML|MR|MU|MW|MZ|NA|NE|NG|RW|SC|SD|SL|SN|SO|SS|ST|SZ|TD|TG|TN|TZ|UG|ZA|ZM|ZW&view=table'
```

```python
chrome_path = "./chromedriver/chromedriver"
service = Service(chrome_path)
chrome_options = Options()
#chrome_options.add_argument("--headless")
#chrome_options.add_argument('--no-sandbox')
#chrome_options.add_argument('--disable-dev-shm-usage')

wd = webdriver.Chrome(service=service,options=chrome_options)
wd.maximize_window() #maximize the window
wd.get(url)          #open the URL
```

```python
wd.implicitly_wait(10000) #maximum time to load the link
html = wd.page_source
```

```python
df = pd.read_html(html)
df1=df[0]
df1
```

```python
wd.close()
```

```python
df1=df1.drop(["Change'10-'19","AAT'10-'19","AAT'15-'19","AT'10-'19","AT'15-'19","Classification"],axis = 1)
df1 = df1.melt(id_vars=['Location'], var_name='year', value_name='value')
df1.rename(columns={ 'Location':'Areaenglish' } ,inplace=True)
```

```python
df1
```

```python
unit=[]
a_code =[]
# Data Source
DataSource='MO Ibrahim Foundation'
# Data Provider ID
DataProviderID=46

for i in range(len(df1)):
    a_code.append(Area_code[df1.iloc[i,0]])
    try:
        df1.iloc[i,1] = int(df1.iloc[i,1])
        unit.append('Rank')
    except:
        df1.iloc[i,1] = int(float(df1.iloc[i,1]))
        unit.append('indx')
mode = ['']*len(df1)
df1.insert(0,'Areacode',a_code)
df1.insert(2,'Unitcode',unit)
df1.insert(2,'Modecode',mode)

cite = f'Retrived from: {url}, On: {str(D.date())}, By web scraping using Selenium and headless chrome web driver'
SpecificSource = [cite]*len(df1)
df1.insert(2, 'SpecificSource', SpecificSource )
df1.insert(2, 'DataProviderId', [DataProviderID]*len(df1) )
df1.insert(2, 'DataSource', [DataSource]*len(df1) )
df1.to_excel('Mo_Ibrahim_Governance_index.xlsx', index=False)
print("All Done!")
```
