# Dictionary for Unit Multipliers

```python
Multiplier={
    'B':['Billions',1000000000],
    'M':['Millions',1000000],
    'TH':['Thousands',1000]
}
```

# Getting all Indicator List

```python
Indicator_List = requests.get('https://unstats.un.org/SDGAPI/v1/sdg/Indicator/List').json()
# Coverting the json to pandas DataFrame
Indicator_List = pd.DataFrame(Indicator_List)
```

```python
# Getting the dictionary that contains all indicators 
IList=[ Indicator_List['series'][i][j] for i in range(len(Indicator_List)) for j in range(len(Indicator_List['series'][i]))]
# Forming DataFrame for the indicatory list
Indicator_List = pd.DataFrame(IList)
```

```python
num = 0
a=1
DataFrameT = pd.DataFrame(columns = ['IndicatorName', 'Indicatorcode','UnUnitcode','Unitcode','Areaenglish','Areacode','DataSource','DataProviderId','SpecificSource','Year','OriginalData','Value'])
D_source = 'United Nations Statistics Division'
D_providerID= 9

for ky in Area_Code.keys():
    geoAreacode = ky
    print(f"Loading Data for {Area_Code[ky][0]}, {Area_Code[ky][1]}")
    # Creating Headers
    DataFrame = pd.DataFrame(columns = ['IndicatorName', 'Indicatorcode','UnUnitcode','Unitcode','Areaenglish','Areacode','DataSource','DataProviderId','SpecificSource','Year','OriginalData','Value'])
    # Loping the indicartor list
    
    
    for j in range(len(Indicator_List)):
        #Getting the series code
        series_code=Indicator_List.loc[j]['code']
        #Getting the goal code for a given series code
        goal_code = Indicator_List.loc[j]['goal'][0]

        #Requesting all dimensions for a given goal
        Dimensions_for_a_Series = None
        while Dimensions_for_a_Series is None:
            try:
                Dimensions_for_a_Series= requests.get(f'https://unstats.un.org/SDGAPI/v1/sdg/Goal/{goal_code}/Dimensions').json()
            except:
                time.sleep(3)
                continue
        D = pd.DataFrame.from_dict(Dimensions_for_a_Series)

        # Requesting data for a given series code  and area code
        Data_req = None
        url = f'https://unstats.un.org/SDGAPI/v1/sdg/Series/Data?seriesCode={series_code}&areaCode={geoAreacode}&pageSize=500'
        while Data_req is None:
            try:
                Data_req = requests.get(url).json()
            except:
                time.sleep(3)
                continue
        Data = pd.DataFrame.from_dict(Data_req['data'])

        # Relating the Data with the dimensions
        S_source= f'Retrived from: {url}, On: {str(Dtd.date())}, By web scraping using the python libraries: requests and pandas'
        for k in range(len(Data)):
            #Getting the unit for a 
            Data_attributes = Data.iloc[k]['attributes']
            Unit= Data_attributes['Units']
            UnmodifiedUnitcode=Unit


            IndicatorCode_base = 'SDG1.'+Data.loc[k]['indicator'][0]+'_'+ series_code
            M = 1
            #Modifiying the Unit into ECA's format and getting the multipliers for the unit
            if Unit not in ECA_Units:
                SplitU = Unit.split('_')
                M_key_l = SplitU[-1]
                M_key_f = SplitU[0]
                if M_key_l in Multiplier.keys():
                    Unit = Unit[:len(Unit)-(len(M_key_l)+1)]
                    M = Multiplier[M_key_l][1]
                elif M_key_f in Multiplier.keys():
                    Unit = Unit[(len(M_key_f)+1):]
                    M = Multiplier[M_key_f][1]
                elif M_key_f=='MT':
                    Unit = Unit[1:]
                    M = Multiplier['M'][1]
                else:
                    pass
            IndicatorCode =IndicatorCode_base
            IndicatorName = Data.loc[k]['seriesDescription']
            #print(IndicatorCode)
            if IndicatorName[-1]==')':
                start = IndicatorName.rfind('(')
                end = IndicatorName.rfind(')')
                last = IndicatorName[start:end+1]
                n = len(last)
                IndicatorName = IndicatorName[:len(IndicatorName)-(n+1)]
                last = f' ({ECA_Units[Unit][1]})'
            else:
                last = ''
            #Dimensions for a given data    
            Data_dimensions=Data.iloc[k]['dimensions']

            # Concatinating the dimensions with the indicator name and code.
            if len(Data_dimensions) !=0:
                for key in sorted(Data_dimensions.keys()):
                    if key not in ['Reporting Type','Freq']:
                        IndicatorCode +='_'+Data_dimensions[key]
                        Dimensions = D[D['id']==key]['codes'].values
                        if len(Dimensions) >0:
                            Dimensions = Dimensions[0]
                            Dim = pd.DataFrame.from_dict(Dimensions)
                            Dimension_description = Dim[Dim['code']==Data_dimensions[key]]['description']
                            D_desc = Dimension_description.values[0]
                            IndicatorName +="//"+D_desc                   
                IndicatorName += last
            #print(IndicatorName)
            #print(IndicatorCode)

            # Cleaning the value: if value is NaN we drop it and if the value contains special characters
            # we remove that character. For example if value="<5" we remove "<" and value will be 5. Once we clean 
            # the value if the unit has a multiplier we multiply the value by the multiplier.
            T=True
            v = Data.iloc[k]['value']
            V=v # Uncleaned value
            #print(v)
            if v=="NaN" or v=="N":
                pass
            else:
                while T:
                    if len(v)>0:
                        try:
                            v= float(v)
                            v *=M
                            if M != 1:
                                v=int(v)
                            # Writing cleaned row data into the DataFrame    
                            DataFrame.loc[len(DataFrame)]=[IndicatorName,IndicatorCode,UnmodifiedUnitcode,ECA_Units[Unit][0],Area_Code[geoAreacode][0],Area_Code[geoAreacode][1],D_source,D_providerID,S_source,Data['timePeriodStart'][k],V, v]
                            T=False
                            break
                        except:
                            v=v[1:]
                    else:
                        v=V
                        # Writing uncleaned row data into the DataFrame    
                        DataFrame.loc[len(DataFrame)]=[IndicatorName,IndicatorCode,UnmodifiedUnitcode,ECA_Units[Unit][0],Area_Code[geoAreacode][0],Area_Code[geoAreacode][1],D_source,D_providerID,S_source,Data['timePeriodStart'][k],V, v]
                        T=False
                        break
        
    num +=1
    if num == 1:
        first_county = Area_Code[ky][0]

    
    if num >10: 
        DataFrameT = pd.concat([DataFrameT, DataFrame], axis=0)
        # Converting the year column into int                    
        DataFrameT['Year'] = DataFrameT['Year'].astype(int)
        # Writing the data to csv
        DataFrameT.to_csv(f'{first_county}_to_{Area_Code[ky][0]}_wide.csv', index=False)
        df = pd.pivot_table(DataFrame, values='Value', columns=['Year'],index=['IndicatorName','Indicatorcode','UnUnitcode','Unitcode','Areaenglish', 'Areacode','DataSource','DataProviderId','SpecificSource'])
        df.to_excel(f'{first_county}_to_{Area_Code[ky][0]}_long.xlsx')
        num=0
        a +=1
        DataFrameT = pd.DataFrame(columns = ['IndicatorName', 'Indicatorcode','UnUnitcode','Unitcode','Areaenglish','Areacode','DataSource','DataProviderId','SpecificSource','Year','OriginalData','Value'])
    else:
        DataFrameT = pd.concat([DataFrameT, DataFrame], axis=0)

    # Converting the year column into int                    
    DataFrame['Year'] = DataFrame['Year'].astype(int)
    # Writing the data to csv
    DataFrame.to_csv(f'{Area_Code[ky][0]}_wide.csv', index=False)
    print(f"\tData for {Area_Code[ky][0]}, {Area_Code[ky][1]}, has been saved!")
    df = pd.pivot_table(DataFrame, values='Value', columns=['Year'],index=['IndicatorName','Indicatorcode','UnUnitcode','Unitcode','Areaenglish', 'Areacode','DataSource','DataProviderId','SpecificSource'])
    df.to_excel(f'{Area_Code[ky][0]}_long.xlsx')
print('All Done!')
```
