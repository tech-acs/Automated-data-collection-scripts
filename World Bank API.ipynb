```python
# Creating Headers
D_source = 'World Bank'
D_providerID= 1
DataFrame = pd.DataFrame(columns = ['IndicatorName', 'IndicatorCode','UnitCode','Areaenglish','Areacode','DataSource','DataProviderId','Year','Value'])
for countryCode in Area_Code:
    
    countryName = Area_Code[countryCode]
    print('Country:', countryName,'(', countryCode,')')
        
    #grab indicators for countires above and load into data frame
    for i in range(len(Indicators_list)):
        ind = Indicators_list[i][0]
        IndicatorName = list(ind.values())[0]
        IndicatorCode1 = list(ind.keys())[0]
        IndicatorCode2 = Indicators_list[i][3]
        unitCode = Indicators_list[i][2]
        print('\t',IndicatorCode2)

        df = None
        a = 0
        while df is None:
            try:
                a +=1
                print("atempt:",a)
                df = wb.data.DataFrame(IndicatorCode1, countryCode,time=range(2000,2019),numericTimeKeys=True)
            except: 
                time.sleep(3)
                continue
        
        D = df.dropna(axis=1)
        
        y = D.columns.values

        V = D.values.T
        v = list(V[:,0])

        indName = [IndicatorName]*len(y)
        indCode = [IndicatorCode2]*len(y)
        cName = [countryName]*len(y)
        cCode = [countryCode]*len(y)
        unitCode = [unitCode]*len(y)
        D_source = [D_source]*len(y)
        D_providerID= [D_source]*len(y)

        D = np.array([indName,indCode,unitCode,cName, cCode,D_source,D_providerID,y,v])

        df1 = pd.DataFrame(D.T, columns=DataFrame.columns)
        DataFrame = pd.concat([DataFrame,df1])
        #DataFrame = DataFrame.concat(df1,ignore_index = True)        


# Converting the year column into int                    
DataFrame['Year'] = DataFrame['Year'].astype(int)
# Writing the data to csv
DataFrame.to_excel('All_Countries_None_Pivoted.xlsx', index=False)
print('All Done')
```
